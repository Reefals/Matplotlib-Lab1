```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([240, 250, 260, 270, 280, 290])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.title("Sports Watch Data", loc = 'center')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.bar(x,y)
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
y = [1,2,3,4]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y,labels=mylabels,  startangle=90,autopct='%1.1f%%', shadow=True,
        radius=1, explode = [0.3,0.2,0,0])

plt.legend(title = "Four Fruits:",loc='upper right')



plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis',alpha = 0.4)

plt.title('colors',fontsize = 20)
plt.xlabel('x',fontsize = 50)
plt.ylabel('y',fontsize =50)

plt.colorbar()

plt.show()




```

## Q4: Add the following to your chart:
